```python
# Chemin du fichier CSV import√© dans Databricks
csv_path = "/Volumes/workspace/default/oc-lakehouse/sales.csv"

df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(csv_path)
)

df.display()
```

```python
%sql
CREATE TABLE IF NOT EXISTS sales_delta
USING DELTA
AS
SELECT
 *
FROM read_files(
  '/Volumes/workspace/default/oc-lakehouse/sales.csv',
  format => 'csv',
  header => true
);
```

```python
%sql
SELECT
  country,
  SUM(quantity * unit_price) AS total_revenue
FROM sales_delta
GROUP BY country
ORDER BY total_revenue DESC;
```
