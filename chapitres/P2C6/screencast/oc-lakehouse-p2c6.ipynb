# Créer un catalogue et un schéma

```sql
CREATE CATALOG IF NOT EXISTS demo_lakehouse;
```

```sql
CREATE SCHEMA IF NOT EXISTS demo_lakehouse.sales
```

```sql
CREATE TABLE IF NOT EXISTS demo_lakehouse.sales.sales_delta_uc
USING DELTA
AS
SELECT
  CAST(order_id AS INT) AS order_id,
  CAST(order_date AS DATE) AS order_date,
  country,
  product,
  CAST(quantity AS INT) AS quantity,
  CAST(unit_price AS DOUBLE) AS unit_price
FROM read_files(
  '/Volumes/workspace/default/oc-lakehouse/sales.csv',
  format => 'csv',
  header => true
);
```
